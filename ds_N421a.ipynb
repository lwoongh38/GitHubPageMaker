## 아래 만들어진 함수를 이용하여 XOR GATE를 제작해보세요. 

```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1

def  NAND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 1 ## only change here
  elif tmp > theta:
    return 0 ## only change here

def OR(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.3 ## only change here
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0 
  elif tmp > theta:
    return 1 
```

이 게이트는 두 입력 값 x1, x2가 다를 때만 1을 출력합니다.

(Hint: 위에서 만든 함수를 종합해서(묶어서) 흐름을 만들도록 활용하면 됩니다)

XOR(1, 1) = 0

XOR(1, 0) = 1

XOR(0, 1) = 1

XOR(0, 0) = 0

---


```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
def XOR1(x1, x2):
    w1, w2 = 0.5, 0.5
    tmp = x1*w1 + x2*w2
    if tmp == 0 or tmp == 1:
        return 0
    elif tmp == 0.5:
        return 1
```

```python
print(XOR1(0,0))
print(XOR1(0,1))
print(XOR1(1,0))
print(XOR1(1,1))
```

```python
def XOR(x1, x2):
    A = NAND(x1, x2)
    B = OR(x1, x2)
    C = AND(A, B)
    return C
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

함수를 작성했는데 문제의도와 다른것 같아 검색.
역시나 오답이었고 왜 답인지를 해석해보았다.

NAND 함수와 OR 함수를 AND 조건으로 묶어주면 NAND와 OR의 결과에서 같은 결과를 낸 (0,1),(1,0)만 1이 출력되고 (0,0),(1,1)은 모두 0이 출력되므로 XOR 문제의 요구사항을 만족한다.

## 신경망에서 용어들에 대해서 자신만의 언어로 설명해보세요. 

인터넷 자료를 찾아도 좋으나, 그대로 쓰지 마시고, 내 언어로 바꾸어 설명해보시는 것이 중요합니다.

먼저 영문 이름을 한글로 바꿔보시고, 영문과 한글이 동시에 머리속에 들어오면 이 과제는 성공입니다. 

image.png

Input Layer: 입력층을 의미하고 입력 변수의 수와 노드의 수는 일치한다. 입력층은 초기에 입력되는 값으로 값들을 바로 은닉층으로 전달만 하고 별다른 계산과정이 없기 때문에 신경망의 층 수를 셀 때 예외로 한다.

Hidden Layer: 은닉층을 의미하고 입력받은 데이터를 전달하는 역할을 한다. 입력층과 출력층 사이의 모든 층을 은닉층이라고 한다.

Output Layer: 출력층. 결과적으로 우리가 얻고자하는 출력 뉴런을 말한다.

image.png

Neuron: 뇌 속에 다수의 뉴런들이 네트워크를 이루며 구성되어 있다. 사진의 실제 인간의 신경망을 보면 여러 감각기관으로부터 전달받은 입력값을 합하여 (임계값에 따라)반응여부를 판단하고 정보를 전달해야 하면 시냅스를 이용하여 다음 뉴런에게 신호를 전달한다.

Weight:가중치를 의미하며 받은 정보에 대하여 중요도를 부여하는 개념이다. 입력받은 정보에 대한 가중치가 높다면 그 정보가 아웃풋을 내는데 결정적인 역할을 한다는 것을 의미한다.

Bias: 뉴런이 활성화에는 외부자극과 별개로 편견도 작용한다. 별다른 자극이 없었더라도 선입견을 가지는 것이 그 예인데 사과같이 생긴 배가 있다면 먹어보지 않는 이상 배를 사과라고 판단할 것이다. 신경망에서도 자극(입력값)이 동일하더라도 편견, **즉 이미 형성되어 있는 시냅스의 연결강도(weight)에 따라 다른 결과를 초래할 수 있다.**

Activation Function:다음 뉴런(노드)로 보낼지 말지를 판단하는 함수를 말한다. 뉴런에서 가중치와 편향을 가지고 계산된 값이 임계치를 넘는지 여부를 통하여 다음 뉴런에 신호전달 여부를 활성화 할지 말지 판단하는 함수.

Node Map:신경망 구조를 시각화하여 나타낸 자료를 말한다.

Perceptron: Perception + Neuron 뉴런이 입력값을 받아서 원리를 해결하는 원리를 따라한 인공뉴런. 입력값에 따라서 가중치를 계산한 후 총 합이 임계값을 넘었을 때 1을 출력, 아닐경우 0을 출력한다.

Epoch: 메모리의 한계, 속도저하 등의 이유 때문에 데이터가 방대한 경우 한꺼번에 집어넣을 수 없다. 그래서 데이터를 나누어주게 되는데 이때 한 번의 epoch는 인공신경망에서 전체 데이터 셋에 대하여 한번 학습을 완료한 상태를 말한다.
